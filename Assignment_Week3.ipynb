# 5.a.Count Plot

```python
plt.figure(figsize=(10,10))
sns.countplot(x = 'SL',data = data1)
sns.set_theme(style="darkgrid")
plt.title("Count of Sepal Lenght ")
plt.show()
```

# 5.b.heatmap on Correlation of features in the data.

```python
plt.figure(figsize=(10,10))
sns.heatmap(data1.corr(),annot = True)
plt.title("Correlation of features ", fontsize=18)
plt.xticks(rotation=90, fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

# 5.c.kernal based density estimation

```python
data1[['SL','SW','PL','PW']].plot.kde()
plt.title("kernal based density estimation of IRIS Flower", fontsize=18)
plt.show()
```

# 5.d.Scatter Plot

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x ='SW', y='PW', hue='Classification', data=data1)
plt.title('Relationship between Sepal Width and Petal Width',fontsize=16)
plt.show()
```
